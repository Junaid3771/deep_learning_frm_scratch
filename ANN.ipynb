```python
import pandas as pd
import numpy as np
```

reading data

```python
data = pd.read_excel(r'clinical_trials.xlsx')
```

```python
data
```

```python
data = data[['Clinical_Trial_Name','Therapeutic_Area']]
```

```python
data.head()
```

tokenization

```python
def tokenize(text):   
    return text.lower().split()


data['clincial_trial_tokenized'] = data['Clinical_Trial_Name'].apply(tokenize)
```

```python
data.head()
```

creating a vocabulary by assigning an id to each word and incrementing it 

```python
vocabulary = {}
id = 0 

for token_list in data['clincial_trial_tokenized']:
    for token in token_list:
        if token not in vocabulary.keys():
            vocabulary[token] = id
            id+=1
            
            
            
```

```python
vocabulary
```

converting the token to ids

```python
def tokens_to_ids(token_list,vocabulary):
    return [vocabulary.get(token) for token in token_list]


data['clinical_trial_id'] = data['clincial_trial_tokenized'].apply(lambda x: tokens_to_ids(x,vocabulary))
```

```python
data
```

applying padding - where we need to ensure the size of each vector is same across rows - we are taking the highest len vector list and applying zeros to the rest to meet this max_length

```python
def padding(id_list, max_len):
    if len(id_list) < max_len:
        id_list = id_list + [0]*(max_len-len(id_list))
    else:
        id_list = id_list[:max_len]
    return id_list


max_len = max(data['clinical_trial_id'].apply(len))

#apply padding 
data['clinical_trial_embedding'] = data['clinical_trial_id'].apply(lambda x: padding(x,max_len))
```

applying hot encoding to theruaptic area column 

```python
def hot_encode(label, classes):
    one_hot = [0]*len(classes)
    index = classes.index(label)
    one_hot[index] = 1
    return one_hot


unique_classes = data['Therapeutic_Area'].unique().tolist()
data['therapeutic_area_embedding'] = data['Therapeutic_Area'].apply(lambda x: hot_encode(x,unique_classes))
```
